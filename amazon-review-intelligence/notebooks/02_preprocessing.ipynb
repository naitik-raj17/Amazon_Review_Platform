Demonstrates the reusable text preprocessing pipeline and sentiment label creation.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from data_loader import prepare_dataset
from preprocess import TextPreprocessor
from utils import load_env

load_env(PROJECT_ROOT / ".env")
```

```python
df = prepare_dataset(PROJECT_ROOT / "data" / "reviews.csv")
print(f"Training-ready rows: {len(df)}")
df[["Rating", "Review Title", "Review Text", "sentiment"]].head()
```

```python
preprocessor = TextPreprocessor()
sample = df.iloc[0]
raw = preprocessor.combine_title_and_text(sample["Review Title"], sample["Review Text"])
cleaned = preprocessor.clean_text(raw)

print("Raw:", raw)
print("Cleaned:", cleaned)
```
