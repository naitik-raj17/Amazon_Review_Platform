# 01 — Exploratory Data Analysis

Explore rating distributions, country analytics, review trends, word frequencies, and review length patterns.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT / "src"))

import matplotlib.pyplot as plt
import pandas as pd
from data_loader import load_dataset, handle_missing_values, remove_duplicates, validate_text_records
from eda import (
    get_dataset_statistics,
    plot_country_ratings,
    plot_rating_distribution,
    plot_review_length_distribution,
    plot_review_trends,
    plot_word_frequency,
)
from preprocess import TextPreprocessor, build_combined_text_column
from utils import load_env, setup_logging

load_env(PROJECT_ROOT / ".env")
logger = setup_logging("notebook_eda")
```

```python
df = load_dataset(PROJECT_ROOT / "data" / "reviews.csv")
df = handle_missing_values(df)
df = remove_duplicates(df)
df = validate_text_records(df)
df = build_combined_text_column(df)
df.head()
```

```python
stats = get_dataset_statistics(df)
stats
```

```python
plot_rating_distribution(df)
plt.show()
```
